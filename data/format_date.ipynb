```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("online_shoppers_intention.csv")
```

```python
df
```

```python
df["Month"].unique()
```

```python
df["Month"].replace("June", "Jun", inplace=True)
```

```python
def convert_date(date_txt):
    dttmp=datetime.strptime(f"{date_txt} 1,2018","%b %d,%Y")
    return dttmp
```

```python
df["Month_Date"] = df["Month"].apply(convert_date)
```

```python
df["Month_Date"]
```

```python
df["Month_Date"] = pd.to_datetime(df["Month_Date"])
```

```python
df.info()
```

```python
df.to_csv("online_shoppers_intention_date_formated.csv", index=False)
```
